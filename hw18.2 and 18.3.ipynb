```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression, LassoCV
from sklearn.model_selection import train_test_split
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar18/sgn.csv')

def regression(m):
    
    plt.tight_layout()
    sns.scatterplot(x=df['x'], y=df['y'])
    
    df1 = df[['x', 'y']]
    sin = ['sin' + str(i) for i in range (1, m+1)]
    cos = ['cos' + str(i) for i in range (1, m+1)]
    
    for i in range(m):
        
        df1[sin[i]]=np.sin(df['x']*i)
        df1[cos[i]]=np.cos(df['x']*i)
    
    X = df1.iloc[:,:-2]
    y = df1['y']
    
    reg = LinearRegression()
    reg.fit(X,y)
    print(reg.coef_)
    print(reg.intercept_)
    
    y_pred = reg.predict(X)
    
    sns.lineplot(x=df['x'], y=y_pred, color = 'Orange')
    
regression(1)
```
